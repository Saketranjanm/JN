```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('mpg')
```

```python
df
```

```python
df.drop("name",axis =1,inplace = True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df['horsepower'].median()
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.origin.unique()
```

```python
df.origin.value_counts()
```

```python
df['origin'] = df['origin'].map({"usa":1,"japan":2,"europe":3})
```

```python
df
```

```python
X = df.drop('mpg',axis = 1)
y = df['mpg']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
regression_model
```

```python
regression_model.fit(X_train,y_train)
```

```python
for i ,col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {regression_model.coef_[i]}")
```

```python
from sklearn.metrics import r2_score
y_pred_linear = regression_model.predict(X_test)
r2_score(y_test,y_pred_linear)
```

```python
#ridge regression
from sklearn.linear_model  import Ridge
ridge_regression_model = Ridge(alpha=0.1)
ridge_regression_model.fit(X_train,y_train)
for i , col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {ridge_regression_model.coef_[i]}")
```

```python
from sklearn.metrics import r2_score
y_pred= ridge_regression_model.predict(X_test)
r2_score(y_test,y_pred)
```

```python
#Lasso regression
from sklearn.linear_model  import Lasso
lasso_regression_model = Lasso(alpha=0.1)
lasso_regression_model.fit(X_train,y_train)
for i , col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {lasso_regression_model.coef_[i]}")
```

```python
from sklearn.metrics import r2_score
y_pred= lasso_regression_model.predict(X_test)
r2_score(y_test,y_pred)
```

```python
#elastic net
from sklearn.linear_model  import ElasticNet
elasticNet_regression_model = ElasticNet(alpha=0.8,l1_ratio=0.4)
elasticNet_regression_model.fit(X_train,y_train)
for i , col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {elasticNet_regression_model.coef_[i]}")
```

```python
from sklearn.metrics import r2_score
y_pred= elasticNet_regression_model.predict(X_test)
r2_score(y_test,y_pred)
```

```python
#cross validation
from sklearn.linear_model import LassoCV
lassocv = LassoCV(cv = 5)
lassocv.fit(X_train, y_train)
y_pred = lassocv.predict(X_test)
r2_score(y_test,y_pred)
```
