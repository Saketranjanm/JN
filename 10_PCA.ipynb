```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
X = data.data
y = data.target
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_trainn = scaler.fit_transform(X_train)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
pca
```

```python
X_trainn = pca.fit_transform(X_train)
```

```python
X_trainn
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_) #more than 90% variance you got, good to go
```

```python
plt.scatter(X_trainn[:,0],X_trainn[:,1],X_trainn[:,2],cmap=plt.cm.nipy_spectral,edgecolors="k")
plt.title("data after pca")
```

```python
plt.scatter(X_train[:,0],X_train[:,1],X_train[:,2],cmap=plt.cm.nipy_spectral,edgecolors="k")
plt.title("original data")
```
